## 19.Расстояние между точками
Сложность: #

Создайте класс Point с одним методом set_coord, который принимает две координаты x и y и сохраняет их в соответствующие атрибуты экземпляра класса.

Также создайте функцию get_dist вне класса, которая принимает на вход два аргумента — точки p1 и p2, являющиеся экземплярами класса Point и вычисляет расстояние между ними.

Координаты для точек p1 и p2 должны передаваться пользователем на вход программы, а расстояние между точками должно быть выведено на экран в результате работы программы.

Подсказка: Помочь вычислить расстояние между точками (x1, y1) и (x2, y2) вам поможет теорема Пифагора. Вспомните, что извлечение корня это эквивалент возведения в степень 0,5.


```python
# Ваш код
class Point:
    def set_coord(self, x, y):
        self.x = x
        self.y = y

def get_dist(p1, p2):
    # Расстояние по теореме Пифагора: √((x2-x1)² + (y2-y1)²)
    return ((p2.x - p1.x) ** 2 + (p2.y - p1.y) ** 2) ** 0.5

# не изменяйте код ниже, он нужен для проверки
p1 = Point()
p2 = Point()
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())
p1.set_coord(x1, y1)
p2.set_coord(x2, y2)
d = get_dist(p1, p2)
print(d)
```

Создайте класс HockeyPlayer, у которого есть:

* инициализатор __init__, принимающий в качестве аргументов first_name и last_name, то есть имя и фамилию хоккеиста. Также во время инициализации объекта необходимо создать два атрибута-счетчика: goals и assists. Эти атрибуты будут хранить информацию о забитых голах и передачах каждого хоккеиста. Оба атрибута необходимо проинициализировать нулями;

* метод add_goals, который добавляет количество голов в счетчик игрока. По умолчанию добавляет один гол;

* метод add_assists, который добавляет количество передач в счетчик игрока. По умолчанию добавляет одну передачу;

* метод statistics, который считает результативность игрока и возвращает строку вида:

  <Имя> <Фамилия> - <голы> + 0.5*<передачи>

Фамилия, имя, количество голов и передач хоккеиста передаются на вход программы пользователем.

```python
# Ваш код
class HockeyPlayer:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name
        self.goals = 0
        self.assists = 0

    def add_goals(self, goals=1):
        self.goals += goals

    def add_assists(self, assists=1):
        self.assists += assists

    def statistics(self):
        return f"{self.first_name} {self.last_name} - {self.goals} + 0.5*{self.assists}"

# не изменяйте код ниже, он нужен для проверки
name = input()
surname = input()
goals = int(input())
assists = int(input())
player = HockeyPlayer(name, surname)
player.add_goals(goals)
player.add_assists(assists)
print(player.statistics())
```

## 24.Класс с Массивом 1
Сложность: #

Создайте класс Array, который содержит атрибут values и 3 метода:

1.   Инициализатор __init__, который принимает произвольное число аргументов, фильтрует их с помощью второго метода и сохраняет в values.

2.   get_integers, который вызывается из инициализатора, фильтрует все его аргументы оставляя только числовые, а затем сортирует их по возрастанию и возвращает в инициализатор в виде списка.

3. Переопределенный метод __str__, который возвращает строку: "Массив(<Значение1>, <Значение2>, ...)", если массив не пустой, или строку "Пустой массив", если массив пуст.

```python
# Ваш код
class Array:
    def __init__(self, *args):
        self.values = self.get_integers(*args)

    def get_integers(self, *args):
        # Фильтруем только числа и сортируем
        numbers = [x for x in args if isinstance(x, (int, float))]
        return sorted(numbers)

    def __str__(self):
        if not self.values:
            return "Пустой массив"
        else:
            # Преобразуем числа в строки и объединяем
            values_str = ", ".join(str(x) for x in self.values)
            return f"Массив({values_str})"

# изменяйте код ниже, он нужен для проверки
v1 = Array(1,"Саша",2,"Вася",3)
print(v1) # печатает "Массив(1, 2, 3)"
v2 = Array()
print(v2) # печатает "Пустой массив"
```

В программе задан класс Array который, в числе прочих, содержит метод __add__. Данный метод принимает на вход числовой массив и  реализует сложение элементов текущего экземпляра класса с соответствующими элементами переданного массива. Доработайте метод, добавив следующие возможности:

* Сложение с типом данных float: каждый элемент массива складывается с переданным дробным числом.

* Если второй аргумент имеет тип данных, отличный от int, float и Array, то выведите сообщение: "Тип данных Array нельзя сложить с <значение 2-го аргумента>"


```python
class Array:
    def __init__(self, *args):
        self.values = Array.get_integers(args)

    @staticmethod
    def get_integers(args):
        arr = [int(value) for value in args if isinstance(value, int)]
        arr.sort()
        return arr

    def __str__(self) -> str:
        if self.values:
            return "Массив(" + str(self.values)[1:-1] + ")"
        return 'Пустой массив'

    def __add__(self, other):
        if isinstance(other, Array):
            if len(self.values) == len(other.values):
                return Array(*[self.values[i] + other.values[i] for i in range(len(self.values))])
            else:
                # Если массивы разной длины, возвращаем пустой массив
                return Array()
        elif isinstance(other, (int, float)):
            # Сложение с числом (int или float)
            return Array(*[x + other for x in self.values])
        else:
            # Неподдерживаемый тип
            print(f"Тип данных Array нельзя сложить с {other}")
            return self  # Возвращаем исходный массив без изменений

# не изменяйте код ниже, он нужен для проверки
v1 = Array(1, 2, 3)
v2 = Array(10, 11, 12)
v3 = Array(4, 5)
print(v1 + v2)
print(v1 + 3)
print(v1 + v3)
print(v3 + 5.91)
```

## 28.Класс с Массивом 3
Сложность: #

В программе задан класс Array. Дополните его двумя магическими методами, которые определяют поведение оператора *. Принцип работы методов:

* Умножение на экземпляр класса Array: поэлементное умножение значений текущего экземпляра класса с соответственными элементами переданного массива.

* Умножение с типом данных int: каждый элемент массива умножается на переданное число. Умножение должно работать вне зависимости от перестановки операндов.

* Если второй аргумент имеет тип данных, отличный от int и Array, то выведите сообщение: "Тип данных Array нельзя перемножить с <значением>".

* Если второй аргумент имеет тип данных Array, но длины массивов не совпадают, то выведите сообщение: "Данные массивы нельзя перемножить, так как в них разное количество элементов!"

Подсказка: Вам нужны методы `__mul__` и` __rmul__`, причем второй просто вызывает первый.

```python
class Array:
    def __init__(self, *args):
        self.values = Array.get_integers(args)

    @staticmethod
    def get_integers(args):
        arr = [int(value) for value in args if isinstance(value, int)]
        arr.sort()
        return arr

    def __str__(self) -> str:
        if self.values:
            return "Массив(" + str(self.values)[1:-1] + ")"
        return 'Пустой массив'

    def __mul__(self, other):
        if isinstance(other, Array):
            if len(self.values) == len(other.values):
                return Array(*[self.values[i] * other.values[i] for i in range(len(self.values))])
            else:
                print("Данные массивы нельзя перемножить, так как в них разное количество элементов!")
                return Array()
        elif isinstance(other, int):
            return Array(*[x * other for x in self.values])
        else:
            print(f"Тип данных Array нельзя перемножить с {other}")
            return self

    def __rmul__(self, other):
        # Умножение с обратным порядком операндов
        return self.__mul__(other)

# не изменяйте код ниже, он нужен для проверки
v1 = Array(1, 2, 3)
v2 = Array(10, 11, 12)
v3 = Array(4)
print(v1 * 2)
print(3 * v2)
print(v2 * v1)
print(v1 * 4.2)
```

```python
class Character:
    def __init__(self, name, health_points, armor_points, weapon=None):
        self.name = name
        self.health_points = health_points
        self.armor_points = armor_points
        self.weapon = weapon

    def __str__(self):
        return f"Name: {self.name}, HP: {self.health_points}, Armor: {self.armor_points}"

    def attack(self, other, hit_points):
        if isinstance(other, Character):
            other.health_points -= hit_points
            print(f"Здоровье {other.name} составляет теперь {other.health_points}")
        else:
            print(f"{other} не является игровым персонажем!")

class Tank(Character):
    def __init__(self, name, health_points, armor_points, weapon="sword", gameclass="Tank"):
        super().__init__(name, health_points, armor_points, weapon)
        self.gameclass = gameclass

    def __str__(self):
        return f"Name: {self.name}, HP: {self.health_points}, Armor: {self.armor_points}, GameClass: {self.gameclass}"

    def massive_disarm(self, *args):
        for character in args:
            if isinstance(character, Character):
                character.armor_points = 0

class Knight(Tank):
    def __init__(self, name, health_points, armor_points, weapon="sword", gameclass="Knight"):
        super().__init__(name, health_points, armor_points, weapon, gameclass)

    def __str__(self):
        return f"Name: {self.name}, HP: {self.health_points}, Armor: {self.armor_points}, GameClass: {self.gameclass}"

    def massive_attack(self, *args):
        for character in args:
            if isinstance(character, Character):
                character.health_points -= 5

class Paladin(Tank):
    def __init__(self, name, health_points, armor_points, weapon=None, gameclass="Paladin"):
        super().__init__(name, health_points, armor_points, weapon, gameclass)

    def __str__(self):
        return f"Name: {self.name}, HP: {self.health_points}, Armor: {self.armor_points}, GameClass: {self.gameclass}"

    def massive_heal(self, *args):
        for character in args:
            if isinstance(character, Character):
                character.health_points += 5

# Не изменяйте код ниже, он служит для проверки
# Базовый класс
c1 = Character('Character_1', 100, 10)
c2 = Character('Character_2', 120, 15)
print(c1.__str__())
print(c2)
c1.attack(c2, 10)
print('-------')

# Класс Tank
t1 = Tank("Красный Дракон", 150, 20)
print(t1)
t1.massive_disarm(c1, c2)  # Снизим броню персонажам c1, c2 до 0
print(c1)  # Проверим
print(c2)
print('-------')

# Класс Knight
k1 = Knight("Полуночный Рыцарь", 95, 25)
print(k1)
k1.massive_attack(c1, c2, t1)  # Нанесем массивный урон по противникам
print(t1)  # Проверим
print(c1)
print(c2)
print('-------')

# Класс Paladin
p1 = Paladin('Паладин Света', 110, 10)
print(p1)
p1.massive_heal(k1)  # Восстановим здоровье союзнику
print(k1)
```

код

```python
class Counter:
    def __init__(self):
        self.value = 0  # Инициализируем счетчик значением 0

    def set(self, start_value=0):
        self.value = start_value

    def increment(self):
        self.value += 1

    def decrement(self):
        self.value -= 1

    def reset(self):
        self.value = 0

    def print(self):
        print(f"Значение счетчика: {self.value}")

n = int(input())

# Ваш код
c = Counter()
c.set(n)  # Отсчет начался с n
c.increment()
c.increment()
c.decrement()
c.print()
c.reset()
c.print()
# Конец вашего кода
```

## **43.Стек**
Сложность: # # #

В этом задании вам предстоить реализовать собственный стек. Стек (англ. stack) — это упорядоченная коллекция элементов, работающая по принципу LIFO (от англ. Last In — First Out, последним пришел — первым вышел). В Python очень легко сделать стек, основанный на списках, поэтому мы выберем списки для основы нашей структуры.

Напишите класс Stack, который имеет:

 1. Метод init для создания нового пустого стека. Этот метод ничего не принимает, но задает атрибут values равным пустому списку;

 2. Метод push() кладет элемент на вершину стека. Этот метод принимает элемент в качестве аргумента функции;

 3. Метод pop() снимает элемент с вершины стека и возвращает его значение;

 4. Метод peek() возвращает значение элемента с вершины стека, но не удаляет его из стека, если стек пустой, то выведите сообщение "Empty Stack";

 5. Метод is_empty() проверяет стек на пустоту, возвращает булево значение;

 6. Метод size() считает количество элементов в стеке и возвращает его.

Пример работы с классом:

код

    s = Stack()
    s.peek()
    print(s.is_empty())
    s.push('cat')  
    s.push('dog')  
    print(s.peek())  
    s.push(True)  
    print(s.size())  
    print(s.is_empty())
    s.push(777)
    print(s.pop())  
    print(s.pop())
    print(s.size())

вывод

    Empty Stack
    True
    dog
    3
    False
    777
    True
    2

Напишите реализацию класса.



```python
class Stack:
    def __init__(self):
        self.values = []

    def push(self, item):
        self.values.append(item)

    def pop(self):
        if not self.is_empty():
            return self.values.pop()

    def peek(self):
        if self.is_empty():
            print("Empty Stack")
        else:
            return self.values[-1]

    def is_empty(self):
        return len(self.values) == 0

    def size(self):
        return len(self.values)

elem1 = input()
elem2 = input()
elem3 = int(input())

# Ваш код
s = Stack()
s.peek()
print(s.is_empty())
s.push(elem1)
s.push(elem2)
print(s.peek())
s.push(True)
print(s.size())
print(s.is_empty())
s.push(elem3)
print(s.pop())
print(s.pop())
print(s.size())
# Конец вашего кода
```
